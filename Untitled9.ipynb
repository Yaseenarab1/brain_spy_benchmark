# **step1**

```python
import scanpy as sc
import anndata as ad
import pandas as pd
import numpy as np
import re
from pathlib import Path
from scipy import sparse
import gc

print("STEP 1 (memory-safe): build merged RNA with top5k genes")

ROOT = Path("/content/drive/MyDrive/state_project")
RAW_DIR = ROOT / "data/raw/GSE174554_RAW"
OUT_DIR = ROOT / "data/anndata"
OUT_DIR.mkdir(parents=True, exist_ok=True)

OUT_FILE = OUT_DIR / "adata_rna_top5k.h5ad"

BAD_PREFIXES = {
    "GSM5319529_SF8963",
}

def is_non_rna(prefix: str) -> bool:
    return "snatac" in prefix.lower()

if OUT_FILE.exists():
    OUT_FILE.unlink()
    print(f"Removed existing {OUT_FILE}")

matrix_files = sorted(RAW_DIR.glob("*_matrix.mtx.gz"))
print(f"Found {len(matrix_files)} matrix files")

ref_gene_ids = None
ref_gene_names = None
gene_sums = None
good_prefixes = []

for mtx_path in matrix_files:
    prefix = mtx_path.name.replace("_matrix.mtx.gz", "")

    if prefix in BAD_PREFIXES:
        print(f"[skip-bad] {prefix}")
        continue
    if is_non_rna(prefix):
        print(f"[skip-nonrna] {prefix}")
        continue

    feat_path = RAW_DIR / f"{prefix}_features.tsv.gz"
    bc_path   = RAW_DIR / f"{prefix}_barcodes.tsv.gz"
    if not feat_path.exists() or not bc_path.exists():
        print(f"[skip-missing] {prefix}: no features/barcodes")
        continue

    try:
        tmp = sc.read_mtx(mtx_path)
    except Exception as e:
        print(f"[skip-fail-mtx] {prefix}: {e}")
        continue

    X = tmp.X
    if not sparse.isspmatrix(X):
        X = sparse.csr_matrix(X)
    X = X.astype(np.float32)

    feats = pd.read_csv(feat_path, header=None, sep="\t")
    if feats.shape[1] >= 2:
        gene_ids   = feats.iloc[:, 0].astype(str).values
        gene_names = feats.iloc[:, 1].astype(str).values
    else:
        gene_ids   = feats.iloc[:, 0].astype(str).values
        gene_names = gene_ids

    if X.shape[0] != len(gene_names):
        print(f"[skip-badshape] {prefix}: X genes={X.shape[0]} vs features={len(gene_names)}")
        continue

    if ref_gene_names is None:
        ref_gene_ids = gene_ids
        ref_gene_names = gene_names
        gene_sums = np.zeros(len(gene_names), dtype=np.float64)
    else:
        if len(gene_names) != len(ref_gene_names):
            print(f"[skip-incompatible] {prefix}: gene list length mismatch")
            continue

    gene_sums += np.asarray(X.sum(axis=1)).ravel()
    good_prefixes.append(prefix)
    print(f"[pass1-ok] {prefix}: cells={X.shape[1]} genes={X.shape[0]}")

    del tmp, X
    gc.collect()

if not good_prefixes:
    raise RuntimeError("No good libraries found in PASS 1. Check RAW_DIR and filters.")

print(f"PASS 1 done. Good libraries: {len(good_prefixes)}")

if gene_sums is None:
    raise RuntimeError("gene_sums is None; PASS 1 failed.")

n_top = 5000
if n_top > len(ref_gene_names):
    n_top = len(ref_gene_names)

top_idx = np.argsort(gene_sums)[-n_top:]
top_idx = np.sort(top_idx)

top_gene_ids = ref_gene_ids[top_idx]
top_gene_names = ref_gene_names[top_idx]

print(f"Selected top {n_top} genes.")
print("Example:", list(top_gene_names[:15]))

gene_indexer = top_idx


adata_rna = None
total_cells = 0

GLOBAL_MAX_CELLS = None

for mtx_path in matrix_files:
    prefix = mtx_path.name.replace("_matrix.mtx.gz", "")

    if prefix not in good_prefixes:
        continue

    feat_path = RAW_DIR / f"{prefix}_features.tsv.gz"
    bc_path   = RAW_DIR / f"{prefix}_barcodes.tsv.gz"

    try:
        tmp = sc.read_mtx(mtx_path)
    except Exception as e:
        print(f"[pass2-skip-fail-mtx] {prefix}: {e}")
        continue

    X_full = tmp.X
    if not sparse.isspmatrix(X_full):
        X_full = sparse.csr_matrix(X_full)
    X_full = X_full.astype(np.float32)

    X_top = X_full[gene_indexer, :]
    X_top = X_top.T.tocsr()

    barcodes = pd.read_csv(bc_path, header=None, sep="\t").iloc[:, 0].astype(str).values
    if X_top.shape[0] != len(barcodes):
        print(f"[pass2-skip-badshape] {prefix}: X_top cells={X_top.shape[0]} vs barcodes={len(barcodes)}")
        del tmp, X_full, X_top
        gc.collect()
        continue

    if GLOBAL_MAX_CELLS is not None:
        remaining = GLOBAL_MAX_CELLS - total_cells
        if remaining <= 0:
            print("Reached GLOBAL_MAX_CELLS cap; skipping remaining libraries.")
            break
        if X_top.shape[0] > remaining:
            idx = np.random.choice(X_top.shape[0], remaining, replace=False)
            idx.sort()
            X_top = X_top[idx, :]
            barcodes = barcodes[idx]

    obs_index = [f"{prefix}_{bc}" for bc in barcodes]
    obs = pd.DataFrame(index=obs_index)

    m = re.search(r"(SF[0-9A-Za-z]+)", prefix)
    sample_id = m.group(1) if m else prefix

    obs["sample_id"] = sample_id
    obs["library"] = prefix
    obs["library_id"] = prefix

    var = pd.DataFrame({"gene_id": top_gene_ids}, index=top_gene_names)
    var.index.name = "gene_symbol"

    adata_lib = ad.AnnData(X=X_top, obs=obs, var=var)

    if not sparse.isspmatrix(adata_lib.X):
        adata_lib.X = sparse.csr_matrix(adata_lib.X, dtype=np.float32)
    else:
        adata_lib.X = adata_lib.X.astype(np.float32)

    if adata_rna is None:
        adata_rna = adata_lib
    else:
        adata_rna = ad.concat(
            (adata_rna, adata_lib),
            join="outer",
            merge="same",
        )

    total_cells += adata_lib.n_obs
    print(f"[merged] {prefix}: now total_cells={total_cells}")

    del tmp, X_full, X_top, adata_lib
    gc.collect()

if adata_rna is None:
    raise RuntimeError("PASS 2 failed: no libraries merged.")

adata_rna.var_names_make_unique()
print("Final concatenated AnnData:", adata_rna)

adata_rna.write(OUT_FILE)
print("Saved:", OUT_FILE)
```

# **step 2 **

```python
import scanpy as sc
import anndata as ad
import pandas as pd
import numpy as np
from pathlib import Path
from scipy import sparse

print("STEP 2 (fixed): QC + normalize + clustering + Tumor/Normal labels, RAM-safe")
print("scanpy:", sc.__version__, "anndata:", ad.__version__)

ROOT = Path("/content/drive/MyDrive/state_project")
ANNDATA_IN  = ROOT / "data/anndata/adata_rna_top5k.h5ad"
ANNDATA_OUT = ROOT / "data/anndata/adata_rna_top5k_step2_preprocessed.h5ad"

TN_META = ROOT / "data/intermediate/GSE174554_Tumor_normal_metadata.txt"
SUP1    = ROOT / "data/intermediate/ii.tsv"

if not ANNDATA_IN.exists():
    raise FileNotFoundError(f"Missing {ANNDATA_IN}. Run Step 1 first.")

adata = sc.read_h5ad(ANNDATA_IN)
print("Loaded:", adata)

if not sparse.isspmatrix(adata.X):
    adata.X = sparse.csr_matrix(adata.X, dtype=np.float32)
else:
    adata.X = adata.X.astype(np.float32)

adata.obs["barcode"] = adata.obs_names.str.split("_").str[-1]

if "sample_id" not in adata.obs.columns:
    src = adata.obs.get("library", pd.Series(adata.obs_names, index=adata.obs_names)).astype(str)
    adata.obs["sample_id"] = src.str.extract(r"(SF[0-9A-Za-z]+)", expand=False)

def attach_tumor_normal_labels(adata, meta_path: Path):
    if not meta_path.exists():
        print(f"{meta_path} not found; skipping Tumor/Normal labels.")
        return False

    print(f"Loading Tumor/Normal metadata from {meta_path}")
    meta = pd.read_csv(meta_path, sep=r"\t|,", engine="python", header=0)
    meta.columns = [c.strip() for c in meta.columns]

    id_col = None
    lab_col = None
    for c in meta.columns:
        cl = c.lower()
        if id_col is None and ("barcode" in cl or "sample#_barcode" in cl or "sample_barcode" in cl):
            id_col = c
        if lab_col is None and ("tumor_normal" in cl or "tumor_normal_annotation" in cl or "annotation" in cl):
            lab_col = c

    if id_col is None and len(meta.columns) >= 1:
        id_col = meta.columns[0]
    if lab_col is None and len(meta.columns) >= 2:
        lab_col = meta.columns[1]

    if id_col is None or lab_col is None:
        print("Could not detect ID/label columns in Tumor_normal_metadata; columns:", meta.columns.tolist())
        return False

    ids = meta[id_col].astype(str)
    parts = ids.str.split("_", n=1, expand=True)
    if parts.shape[1] != 2:
        print("Unexpected ID format in Tumor_normal_metadata (expected 'SFxxxxx_BARCODE'). Example:", ids.head().tolist())
        return False

    meta["sample_id"] = parts[0]
    meta["barcode"]   = parts[1]
    meta["author_cell_label"] = meta[lab_col].astype(str)

    def resolve_labels(labels):
        uniq = set(labels)
        if len(uniq) == 1:
            return labels.iloc[0]
        if "Tumor" in uniq:
            return "Tumor"
        if "tumor" in uniq:
            return "Tumor"
        return list(uniq)[0]

    meta_agg = (
        meta.groupby(["sample_id", "barcode"], as_index=True)["author_cell_label"]
        .apply(resolve_labels)
    )

    print(f"Unique (sample_id, barcode) entries in metadata: {meta_agg.shape[0]}")

    keys = list(zip(adata.obs["sample_id"], adata.obs["barcode"]))
    matched = [meta_agg.get(k, np.nan) for k in keys]

    adata.obs["author_cell_label"] = matched

    ser = pd.Series(matched, dtype="object")
    n_matched = ser.notna().sum()
    print(f"Attached Tumor/Normal labels to {n_matched} cells.")

    if n_matched == 0:
        print("WARNING: 0 matches. Check sample_id and barcode parsing.")
        return False

    print("author_cell_label value_counts (non-null):")
    print(adata.obs["author_cell_label"].value_counts(dropna=True))
    return True

has_author_labels = attach_tumor_normal_labels(adata, TN_META)

if SUP1.exists():
    print("Loading sample-level metadata from", SUP1)
    sup = pd.read_csv(SUP1, sep=r"\t|,", engine="python")
    sup.columns = [c.strip() for c in sup.columns]

    sample_col = None
    cond_col = None
    for c in sup.columns:
        cl = c.lower()
        if sample_col is None and ("sample_id" in cl or cl == "sample"):
            sample_col = c
        if cond_col is None and any(w in cl for w in ["primary", "recurrent", "condition"]):
            cond_col = c

    if sample_col is not None and "sample_id" in adata.obs.columns:
        sup["sample_id"] = sup[sample_col].astype(str)
        if cond_col is not None:
            sup["condition"] = (
                sup[cond_col]
                .astype(str)
                .str.lower()
                .map({"primary": "primary", "recurrent": "recurrent"})
            )
        else:
            sup["condition"] = np.nan

        sample_to_cond = dict(zip(sup["sample_id"], sup["condition"]))
        adata.obs["condition"] = adata.obs["sample_id"].map(sample_to_cond)
        print("condition counts (from SupTable1):")
        print(adata.obs["condition"].value_counts(dropna=False))
    else:
        print("Could not map SupTable1 to adata; missing sample_id.")
else:
    print("SupTable1 not found; skipping primary/recurrent labels.")

print("Calculating QC metrics")
sc.pp.calculate_qc_metrics(adata, inplace=True)

min_counts = 200
min_genes = 100

before = adata.n_obs
adata = adata[
    (adata.obs["total_counts"] >= min_counts)
    & (adata.obs["n_genes_by_counts"] >= min_genes)
].copy()
after = adata.n_obs
print(f"Filtered low-quality cells: {before} -> {after}")

mt_genes = [g for g in adata.var_names if g.upper().startswith("MT-")]
if mt_genes:
    mito = np.asarray(adata[:, mt_genes].X.sum(axis=1)).ravel()
    adata.obs["pct_counts_mito"] = mito / adata.obs["total_counts"].values * 100.0
    before = adata.n_obs
    adata = adata[adata.obs["pct_counts_mito"] < 20].copy()
    after = adata.n_obs
    print(f"Filtered high-mito cells: {before} -> {after}")

print(" Shape after QC:", adata.shape)

MAX_CELLS = 80000

if adata.n_obs > MAX_CELLS:
    print(f"Downsampling from {adata.n_obs} to {MAX_CELLS} cells for RAM safety.")
    idx = np.random.choice(adata.n_obs, MAX_CELLS, replace=False)
    idx.sort()
    adata = adata[idx, :].copy()

print("After downsampling (if applied):", adata.shape)

print(" Normalize + log1p")
sc.pp.normalize_total(adata, target_sum=1e4)
sc.pp.log1p(adata)

print("HVGs (flavor='seurat' to avoid skmisc)")
sc.pp.highly_variable_genes(
    adata,
    n_top_genes=2000,
    subset=True,
    flavor="seurat"
)
print("HVG subset:", adata.shape)

print("Scale")
sc.pp.scale(adata, max_value=10)

print("PCA")
sc.tl.pca(adata, n_comps=50, svd_solver="randomized")

print("Neighbors")
sc.pp.neighbors(adata, n_neighbors=15, n_pcs=30)

print("UMAP")
sc.tl.umap(adata)

print("Leiden (0.5)")
sc.tl.leiden(adata, resolution=0.5, key_added="leiden_0_5")
print(adata.obs["leiden_0_5"].value_counts())

def map_author_to_major(label: str) -> str:
    if not isinstance(label, str) or label is None or label != label:
        return "unknown"
    x = str(label).lower()
    if "tumor" in x:
        return "tumor"
    if "normal" in x:
        return "non_tumor"
    return "unknown"

if has_author_labels and "author_cell_label" in adata.obs:
    print("Using Tumor_Normal_annotation as major_cell_type basis.")
    adata.obs["major_cell_type"] = pd.Categorical(
        adata.obs["author_cell_label"].apply(map_author_to_major)
    )
else:
    print("No Tumor/Normal labels attached; using simple tumor-like marker rule.")
    def score_genes(ad, genes):
        genes = [g for g in genes if g in ad.var_names]
        if not genes:
            return np.zeros(ad.n_obs)
        return np.asarray(ad[:, genes].X.mean(1)).ravel()
    tumor_m = score_genes(adata, ["SOX2", "OLIG2", "PDGFRA", "EGFR"])
    labels = np.where(tumor_m > 0.25, "tumor", "non_tumor")
    adata.obs["major_cell_type"] = pd.Categorical(labels)

print("major_cell_type counts:")
print(adata.obs["major_cell_type"].value_counts(dropna=False))

adata.write(ANNDATA_OUT)
print("STEP 2 complete. Saved:", ANNDATA_OUT)
```

```python
import scanpy as sc
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

ROOT = Path("/content/drive/MyDrive/state_project")
IN_FILE = ROOT / "data/anndata/adata_rna_top5k_step2_preprocessed.h5ad"

print("Loading preprocessed AnnData for visualization")
adata = sc.read_h5ad(IN_FILE)
print(adata)

ad_plot = adata

if "X_umap" in ad_plot.obsm_keys():
    print("UMAP: leiden_0_5")
    if "leiden_0_5" in ad_plot.obs:
        sc.pl.umap(
            ad_plot,
            color="leiden_0_5",
            legend_loc="on data",
            frameon=False,
            s=5,
        )

    if "author_cell_label" in ad_plot.obs:
        print("UMAP: Tumor vs Normal (author_cell_label)")
        sc.pl.umap(
            ad_plot,
            color="author_cell_label",
            frameon=False,
            s=5,
            legend_loc="right",
        )

    if "major_cell_type" in ad_plot.obs:
        print("UMAP: major_cell_type")
        sc.pl.umap(
            ad_plot,
            color="major_cell_type",
            frameon=False,
            s=5,
            legend_loc="right",
        )

    if "condition" in ad_plot.obs:
        print("UMAP: condition (primary vs recurrent if available)")
        sc.pl.umap(
            ad_plot,
            color="condition",
            frameon=False,
            s=5,
            legend_loc="right",
        )
else:
    print("No UMAP found in .obsm['X_umap']; did Step 2 finish fully?")

print("QC histograms")
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

axes[0].hist(ad_plot.obs["total_counts"], bins=50)
axes[0].set_title("Total counts per cell")
axes[0].set_xlabel("Total counts")
axes[0].set_ylabel("Number of cells")

axes[1].hist(ad_plot.obs["n_genes_by_counts"], bins=50)
axes[1].set_title("Detected genes per cell")
axes[1].set_xlabel("n_genes_by_counts")

plt.tight_layout()
plt.show()

if "author_cell_label" in ad_plot.obs and "sample_id" in ad_plot.obs:
    print("Tumor/Normal composition per sample")

    comp = (
        ad_plot.obs
        .dropna(subset=["author_cell_label"])
        .groupby(["sample_id", "author_cell_label"], observed=False)
        .size()
        .reset_index(name="n")
    )

    if not comp.empty:
        comp["frac"] = comp.groupby("sample_id")["n"].transform(lambda x: x / x.sum())
        pivot = comp.pivot(index="sample_id",
                           columns="author_cell_label",
                           values="frac").fillna(0)

        pivot = pivot.astype(float)

        if pivot.shape[0] > 0 and pivot.shape[1] > 0:
            pivot = pivot.sort_index()
            pivot.plot(
                kind="bar",
                stacked=True,
                figsize=(12, 4),
            )
            plt.ylabel("Fraction of cells")
            plt.title("Tumor vs Normal composition per sample")
            plt.xticks(rotation=90)
            plt.tight_layout()
            plt.show()
        else:
            print("Pivot table for sample composition is empty; skipping plot.")
    else:
        print("No non-null author_cell_label entries; skipping sample composition plot.")
else:
    print("Missing author_cell_label or sample_id; skipping Tumor/Normal per sample plot.")

if "author_cell_label" in ad_plot.obs and "condition" in ad_plot.obs:
    print("Tumor/Normal composition per condition")

    comp2 = (
        ad_plot.obs
        .dropna(subset=["author_cell_label", "condition"])
        .groupby(["condition", "author_cell_label"], observed=False)
        .size()
        .reset_index(name="n")
    )

    if not comp2.empty:
        comp2["frac"] = comp2.groupby("condition")["n"].transform(lambda x: x / x.sum())
        pivot2 = comp2.pivot(index="condition",
                             columns="author_cell_label",
                             values="frac").fillna(0)
        pivot2 = pivot2.astype(float)

        if pivot2.shape[0] > 0 and pivot2.shape[1] > 0:
            pivot2.plot(
                kind="bar",
                stacked=True,
                figsize=(6, 4),
            )
            plt.ylabel("Fraction of cells")
            plt.title("Tumor vs Normal by condition")
            plt.xticks(rotation=0)
            plt.tight_layout()
            plt.show()
        else:
            print("Pivot table for condition composition is empty; skipping plot.")
    else:
        print(" No valid (condition, author_cell_label) pairs; skipping condition composition plot.")
else:
    print("Missing author_cell_label or condition; skipping Tumor/Normal per condition plot.")
```

# **step 3**

```python
import scanpy as sc
import anndata as ad
import pandas as pd
import numpy as np
from pathlib import Path

print("STEP 3: Build spatial AnnData (aligned to snRNA)")

ROOT = Path("/content/drive/MyDrive/state_project")

ANNDATA_DIR = ROOT / "data/anndata"
INTER_DIR   = ROOT / "data/intermediate"
ANNDATA_DIR.mkdir(parents=True, exist_ok=True)
INTER_DIR.mkdir(parents=True, exist_ok=True)

RNA_FILE = ANNDATA_DIR / "adata_rna_top5k_step2_preprocessed.h5ad"

SP_TXT_GZ  = ROOT / "data/intermediate/GSE174554_Spatial_transcriptomic_matrix.txt.gz"
SP_TXT     = ROOT / "data/intermediate/GSE174554_Spatial_transcriptomic_matrix.txt"
SP_OUT     = ANNDATA_DIR / "adata_spatial_step3_aligned.h5ad"

SAMPLE_ANN = INTER_DIR / "sample_annotations.csv"
SUP1       = INTER_DIR / "SupTable1_samples.tsv"

if not RNA_FILE.exists():
    raise FileNotFoundError(f"Missing {RNA_FILE}. Run Step 2 before Step 3.")

adata_rna = sc.read_h5ad(RNA_FILE)
print("Loaded RNA:", adata_rna.shape)
rna_genes = set(adata_rna.var_names.astype(str))

if SP_TXT.exists():
    sp_path = SP_TXT
elif SP_TXT_GZ.exists():
    sp_path = SP_TXT_GZ
else:
    raise FileNotFoundError(
        "Could not find GSE174554_Spatial_transcriptomic_matrix.txt(.gz). "
        "Put it under data/intermediate/."
    )

print("Loading spatial transcriptomic matrix from", sp_path)

st = pd.read_csv(sp_path, sep=r"\s+|\t", engine="python")
print("Raw spatial table shape:", st.shape)

gene_col = st.columns[0]
st = st.rename(columns={gene_col: "gene"})
st["gene"] = (
    st["gene"]
    .astype(str)
    .str.strip()
    .str.replace('"', "", regex=False)
)

st = st.set_index("gene")

st.columns = (
    st.columns.astype(str)
    .str.strip()
    .str.replace('"', "", regex=False)
)

print("Spatial matrix (genes x spots):", st.shape)

sp_genes = set(st.index.astype(str))
common_genes = sorted(rna_genes & sp_genes)

print(f" Overlap genes between snRNA (5k) and spatial matrix: {len(common_genes)}")

if len(common_genes) == 0:
    gene_sums = st.sum(axis=1).astype(float)
    top_genes = gene_sums.sort_values(ascending=False).head(2000).index.tolist()
    st_use = st.loc[top_genes]
else:
    st_use = st.loc[common_genes]

print(" Using genes for spatial AnnData:", st_use.shape[0])

X = st_use.T.values
spots = st_use.columns.astype(str).tolist()
genes = st_use.index.astype(str).tolist()

obs = pd.DataFrame(index=spots)
obs["spot_id"] = spots

obs["sample_id"] = obs["spot_id"].str.extract(r"(SF[0-9A-Za-z]+)", expand=False)

var = pd.DataFrame(index=genes)
var.index.name = "gene_symbol"

adata_sp = ad.AnnData(X=X, obs=obs, var=var)
print("  Built spatial AnnData:", adata_sp)

sample_to_condition = {}

if SAMPLE_ANN.exists():
    ann = pd.read_csv(SAMPLE_ANN)
    cols = [c.lower() for c in ann.columns]
    if "sample_id" in [c.lower() for c in ann.columns]:
        sid_col = ann.columns[[c.lower() == "sample_id" for c in ann.columns]][0]
    else:
        sid_col = None

    cond_col = None
    for c in ann.columns:
        cl = c.lower()
        if "condition" in cl or "primary" in cl or "recurrent" in cl:
            cond_col = c
            break

    if sid_col is not None and cond_col is not None:
        tmp = ann[[sid_col, cond_col]].copy()
        tmp.columns = ["sample_id", "condition_raw"]
        tmp["condition"] = (
            tmp["condition_raw"]
            .astype(str)
            .str.lower()
            .map({"primary": "primary", "recurrent": "recurrent"})
        )
        sample_to_condition = dict(
            tmp.dropna(subset=["condition"])[["sample_id", "condition"]].values
        )
        print(f" {len(sample_to_condition)} sample conditions from sample_annotations.csv")

if not sample_to_condition and SUP1.exists():
    sup = pd.read_csv(SUP1, sep=r"\t|,", engine="python", header=None)
    sup = sup.rename(columns={0: "sample_id", 1: "Condition_guess"})
    sup["sample_id"] = sup["sample_id"].astype(str)
    sup["condition"] = (
        sup["Condition_guess"]
        .astype(str)
        .str.lower()
        .map({"primary": "primary", "recurrent": "recurrent"})
    )
    sample_to_condition = dict(
        sup.dropna(subset=["condition"])[["sample_id", "condition"]].values
    )
    print(f"Loaded {len(sample_to_condition)} sample conditions from SupTable1_samples.tsv")

if sample_to_condition:
    adata_sp.obs["condition"] = adata_sp.obs["sample_id"].map(sample_to_condition)
    print("condition counts in spatial")
    print(adata_sp.obs["condition"].value_counts(dropna=False))

print("calculating QC ")
sc.pp.calculate_qc_metrics(adata_sp, inplace=True)

before = adata_sp.n_obs
adata_sp = adata_sp[adata_sp.obs["total_counts"] >= 10, :].copy()
after = adata_sp.n_obs
print(f" Filtered low-count spots: {before} -> {after}")

print(" Final spatial AnnData:", adata_sp)

adata_sp.write(SP_OUT)
print(" STEP 3 done. Saved:", SP_OUT)
```
